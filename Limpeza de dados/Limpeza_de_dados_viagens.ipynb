## Índices

Definindo indice para o ordenação

```python
df = df.set_index('Identificador do processo de viagem')
df = df.sort_index()

df
```

# Verificando informações do arquivo e conferindo se o Dtype esta correto

```python
df.info()
```

Convertendo colunas do tipo String para seus valores corretos, como data e numerico (Não consegui converter Object para numerico).

```python
df['Período - Data de fim'] = pd.to_datetime(df['Período - Data de fim'])
df['Período - Data de início'] = pd.to_datetime(df['Período - Data de início'])

# for valores in ['Valor diárias','Valor passagens','Valor devolução','Valor outros gastos']:
#   df[valores] = pd.to_numeric(df[valores], errors='ignore')

df.info()
```

Apresenta linhas com NaN no arquivo 

```python
df[df.isna().any(axis=1)]
```

Subistituindo valores nulos por "Não infomado"

```python
df['CPF viajante'] = df['CPF viajante'].fillna("Não informado") 
df['Cargo'] = df['Cargo'].fillna("Não informado")
df['Descrição Função'] = df['Descrição Função'].fillna("Não informado")
df['Justificativa Urgência Viagem'] = df['Justificativa Urgência Viagem'].fillna("Não informado")

df[df.isna().any(axis=1)]
```

Salvando os dados limpos no arquivo

```python
df.to_csv('/content/drive/MyDrive/UP/Ciência de Dados/Notebooks/data/2022_Viagem.csv', index=False)
```
